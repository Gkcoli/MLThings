# Extract Luxury Brands

```python
def extract_luxury(brand):
   luxury_yes = [
   'Lexus','Audi','BMW','Mercedes-Benz','Jaguar',
   'Porsche','Rolls-Royce','Lamborghini','Maserati',
   'Ferrari', 'McLaren','Bentley','Tesla', 'Cadillac',
   'McLaren','Volvo','Land Rover', 'Genesis',
]
   Yes = "Yes"
   No = "No"
   for is_luxury in luxury_yes:
       if is_luxury.lower() in brand.lower():
           return Yes
   return No
   df['is_luxury'] = df['brand'].apply(extract_luxury)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
y_pred_train = linear_reg.predict(X_train)
y_pred_test = linear_reg.predict(X_test)
```

```python
#Import the necessary functions from sklearn
from sklearn.metrics import mean_squared_error, r2_score

train_mse = mean_squared_error(y_train, y_pred_train)
train_r2 = r2_score(y_train, y_pred_train)


test_mse = mean_squared_error(y_test, y_pred_test)
test_r2 = r2_score(y_test, y_pred_test)

print(f"Training MSE:{train_mse}")
print(f"Testing MSE:{test_mse}")
print(f"Training R2 Score:{train_r2}")
print(f"Testing R2 Score:{test_r2}")
```
