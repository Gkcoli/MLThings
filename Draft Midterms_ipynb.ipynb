# Extract Luxury Brands

```python
def extract_luxury(brand):
  luxury_brands = [
      'Lexus', 'Audi', 'BMW', 'Mercedes-Benz', 'Jaguar', 'Porsche', 'Rolls-Royce', 'Lamborghini',
      'Maserati', 'Ferrari', 'McLaren', 'Bentley', 'Tesla', 'Cadillac', 'Volvo', 'Land Rover', 'Genesis'
  ]
  return "Yes" if any(lux_brand.lower() in brand.lower() for lux_brand in luxury_brands) else "No"

df['is_luxury'] = df['brand'].apply(extract_luxury)
```
